### 4. Convert per baris pada text file menjadi beberapa list dalam list

```python
a_file = open("LB_TA_2.txt", "r")

list_of_lists = []
for line in a_file:
  stripped_line = line.strip()
  line_list = stripped_line.split()
  list_of_lists.append(line_list)

a_file.close()

list_of_lists
```

### 5. hapus elemen pada list yang tidak bertipe data float dan string

```python
def is_float(element):
    try:
        float(element)
    except ValueError:
        return False
    return True

test_list = list_of_lists
  
print("The original list : ")
test_list


res = [[x for x in sub if is_float(x)] and [x for x in sub if str] for sub in test_list]
  
res
```

### 6. hapus beberapa list yang kosong yang terdapat pada list tersebut

```python
list2 = [x for x in res if x != []]
list2
```

### 7. Buat Dictionary dari list sebelumnya

```python
s = list2
data_dict = {}
for x in s:
    data_dict.setdefault(x[0],[]).append(x[1])
data_dict
```

### 8. konversi tipe data values setiap dictionary menjadi float

```python
d = data_dict
dict={}
for i in d:
    dict[i]=[]
    for j in d[i]:
        dict[i].append(float(j))


print(dict)
```
