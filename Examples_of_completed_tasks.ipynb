```python
import random

Min = 1
Max = 6
num = int(input('Number of rolling : '))
total = 0

for i in range(num) :
    print('Rolling the dices ...')
    print('The values are ...')
    
    dice1 = random.randint(Min, Max)
    print("Dice 1 = ", dice1)
    dice2 = random.randint(Min, Max)
    print("Dice 2 = ", dice2)
    
    i = i + 1
    total = total + dice1 + dice2

print("\n")
print("Number of rolling is ", i)
print("Total value of Dice 1 and Dice 2 is ", total)
print("Average value of two dices rolling is ", (total/i))
```

```python
# Vegetable = Rp5000/kg         --> Diskon 10%, minimal pembelian 2kg
# Drinks    = Rp10000/kaleng    --> Diskon 8%, minimal pembelian 6 kaleng
# Eggs      = Rp20000/kg        --> Diskon 15%, minimal pembelian 5kg
# Meats     = Rp75000/kg        --> Diskon 10%, minimal pembelian 3kg

# >>>>>>>>>> SAYUR <<<<<<<<<<<
print("\n========= Harga Sayur = Rp5.000 =========")
print("========= Diskon 10% , minimal pembelian 2kg =========")
v = 5000
veg = int(input('Jumlah sayur : '))
if veg >= 2:
    hs = veg*v      #harga sayur
    print("DISKON sayur = Rp",(hs*0.1))
    d1 = (hs*0.1)   #diskon yang didapat
    print("Harga sayur yang dibeli = Rp",hs - d1)
else :
    veg*v
    print("Harga sayur = ",(veg*v))


# >>>>>>>>>> MINUMAN <<<<<<<<<<<
print("\n========= Harga Minuman = Rp10.000 =========")
print("========= Diskon 8% , minimal pembelian 6 kaleng =========")
d = 10000
mnm = int(input('Jumlah minuman : '))
if mnm >= 6:
    hm = mnm*d      #harga minuman
    print("DISKON minuman = Rp",(hm*0.08))
    d2 = (hm*0.08)  #diskon yang didapat
    print("Harga minuman yang dibeli = Rp",hm - d2)
else :
    mnm*d
    print("Harga minuman yang dibeli = ",(mnm*d))


# >>>>>>>>>> TELUR <<<<<<<<<<<
print("\n========= Harga TELUR = Rp20.000 =========")
print("========= Diskon 15% , minimal pembelian 5kg =========")
e = 20000
tel = int(input('Jumlah telur : '))
if tel >= 15:
    ht = tel*e      #harga telur
    print("DISKON telur = Rp",(ht*0.08))
    d3 = (ht*0.08)  #diskon yang didapat
    print("Harga telur yang dibeli = Rp",ht - d3)
else :
    tel*e
    print("Harga telur yang dibeli = ",(tel*e))


# >>>>>>>>>> DAGING <<<<<<<<<<<
print("\n========= Harga daging = Rp75.000 =========")
print("========= Diskon 10% , minimal pembelian 3kg =========")
m = 75000
dag = int(input('Jumlah daging : '))
if dag >= 3:
    hd = tel*m      #harga daging
    print("DISKON daging = Rp",(hd*0.08))
    d4 = (hd*0.08)  #diskon yang didapat
    print("Harga daging yang dibeli = Rp",hd - d4)
else :
    dag*m
    print("Harga daging yang dibeli = Rp ",(dag*m))
```
